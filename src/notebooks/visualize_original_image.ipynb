```python
import glob
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import sys
from torch.utils.data import DataLoader

sys.path.append("../")
from dataset import CigaretteLighterDataset
```

```python
data_dir = '/workspace/src/data/cigarette_lighter/trim/'
img = np.array(Image.open(data_dir+'0001.jpeg'))
plt.imshow(img)
plt.show()
```

```python
file_path = glob.glob(data_dir+'*')
```

```python
original_data = CigaretteLighterDataset(
    mode = 'generate',
    x = file_path
)
original_loader = DataLoader(
    original_data,
    batch_size = 1,
    num_workers = 1,
    pin_memory = False,
    drop_last = False
)
```

```python
fig = plt.figure(figsize=(4, 4))
for index, image in enumerate(original_loader):
    image = image.numpy()
    plt.subplot(4, 4, index+1)
    plt.imshow((image[0].transpose(1, 2, 0) * 127.5 + 127.5).astype(np.uint8))
    plt.axis('off')
    if index == 15:
        break
plt.show()
```
